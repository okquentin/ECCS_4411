```coconut
def factorial(n):
    """Compute n! where n is an integer >= 0."""
    if n `isinstance` int and n >= 0:
        acc = 1
        for x in range(1, n+1):
            acc *= x
        return acc
    else:
        raise TypeError("the argument to factorial must be an integer >= 0")

# Test cases:
cases = [-1, 0.5, 0, 3]
for case in cases:
    try:
        result = factorial(case)
        print(result)
    except TypeError as e:
        print(e)
```
